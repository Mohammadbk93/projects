````python
import requests
from twilio.rest import Client

STOCK_NAME = "TSLA"
COMPANY_NAME = "Tesla Inc"

STOCK_ENDPOINT = "https://www.alphavantage.co/query"
NEWS_ENDPOINT = "https://newsapi.org/v2/everything"

STOCK_API_KEY= "..."
NEWS_API_KEYS = "..."
TWILIO_SID = "..."
TWILIO_AUTH_TOKEN = "..."
````

````python
# When stock price increase/decreases by 5% between yesterday and the day before yesterday then print("Get News").
stock_params = {
    "function": "TIME_SERIES_DAILY",
    "symbol": STOCK_NAME,
    "apikey":STOCK_API_KEY,
}

response = requests.get(STOCK_ENDPOINT, params=stock_params)
print(response.json())
````

### It's a large dictionary


````python
data = response.json()["Time Series (Daily)"]
print(data)
````

### Turning this large dictionary into a List

Creating a new list as (List Comprehension)

```
# new_list = [new_item for items in list]

```





````python
data_list =[value for (key,value) in data.items()]
data_list
````

````python
# Getting yesterday's closing stock price

yesterday_data = data_list[0]
yesterday_data_closing_price = yesterday_data["4. close"]
yesterday_data_closing_price
````

````python
# Getting the day before yesterday's closing stock price

day_before_yesterday_data = data_list[1]
day_before_yesterday_closing_price = day_before_yesterday_data["4. close"]
day_before_yesterday_closing_price
````

````python
#Finding the positive difference between 1 and 2. e.g. 40 - 20 = -20, but the positive difference is 20.

difference = abs(float(yesterday_data_closing_price) - float(day_before_yesterday_closing_price))
difference
````

````python
# Working out the percentage difference in price between closing price yesterday and closing price the day before yesterday.
diff_percent = (difference / float(yesterday_data_closing_price)) * 100
diff_percent
````

````python
# If percentage is greater than 2 then print("Get News").
if abs(diff_percent) > 2:
  print("Get News")
else:
  print("No News")
````

````python
# Instead of printing ("Get News"), we use the News API to get articles related to the COMPANY_NAME.
if diff_percent > 2:
  news_params = {
      "apikey" : NEWS_API_KEYS,
      "q" : COMPANY_NAME,
}
news_response = requests.get(NEWS_ENDPOINT, params = news_params)
articles = news_response.json()["articles"]
articles
````

````python
#  Using Python slice operator to create a list that contains the first 3 articles.

three_articles = articles[:3]
three_articles
````

````python
## Using twilio.com/docs/sms/quickstart/python
#to send a separate message with each article's title and description to your phone number.

# Creating a new list of the first 3 article's headline and description using list comprehension.
formated_articles = [f"Headline: {article['title']}. \nBrief: {article['description']}" for article in three_articles]
formated_articles
````

````python
# Sending each article as a separate message via Twilio.
client = Client(TWILIO_SID, TWILIO_AUTH_TOKEN)
for article in formated_articles:
    message = client.messages.create(
      body = article,
      from_ = "+17542128979",
      to = "your phone number can be placed here"
)
````
